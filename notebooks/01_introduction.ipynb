  1.1 Overview
</h4>

This project analyzes the Bank Marketing dataset from the UCI Machine Learning Repository, which contains the results of direct marketing campaigns conducted via phone calls by a Portuguese banking institution.


The primary goal is to predict whether a client will subscribe to a term deposit based on demographic, financial, and campaign-related information available before the marketing contact. The analysis follows a structured data science workflow including data understanding, exploratory data analysis (EDA), data preparation, supervised modeling, and evaluation.

">
  1.2 Business Context & Objective
</h4>

Direct marketing campaigns are widely used by banks to promote financial products. However, such campaigns are expensive and can negatively impact customer satisfaction if they are poorly targeted.

The objective of this project is to support marketing decision-making by identifying clients who are more likely to subscribe to a term deposit. Accurate predictions can help banks:
- Reduce unnecessary customer contacts
- Allocate marketing resources more efficiently
- Improve campaign success rates

The resulting model is intended as a decision-support tool rather than an automated decision system.

">
  1.3 Dataset Description
</h4>

The dataset used in this project is the Bank Marketing dataset, originally published in the UCI Machine Learning Repository. It contains information from direct marketing phone call campaigns conducted between 2008 and 2010.

Each observation represents a single contact between the bank and a client. The full dataset version is used, containing:
- 41,188 observations
- 20 input features
- 1 binary target variable (y), indicating whether the client subscribed to a term deposit

The features include client demographics, campaign-related attributes, and macroeconomic indicators.

  1.5 Target Leakage Consideration
</h4>

Although the feature duration is highly predictive of the target variable, it is only known after a phone call has been completed. Including this variable would therefore introduce target leakage, resulting in an unrealistically optimistic model performance.

For this reason, duration is excluded from all predictive modeling, ensuring that the models reflect a realistic real-world deployment scenario.

<h4 style="
  margin-bottom: 4px;

">
  1.6 Problem Formulation
</h4>

The task is formulated as a binary classification problem, where the objective is to predict whether a client will subscribe to a term deposit (`yes` or `no`) based solely on information available prior to the marketing contact.

<h4 style="
  margin-bottom: 4px;
  background-color: #f3f4f6;

  border-radius: 4px;
  display: inline-block;
  color: black
">
  1.7 Assumptions, Scope & Ethics
</h4>

This analysis assumes that the dataset is representative of future marketing campaigns. Temporal dependencies are not explicitly modeled.

The dataset contains sensitive demographic attributes such as age, occupation, and marital status, which may introduce bias. Therefore:
- The model is treated as a support tool
- Results are interpreted critically
- No individual-level decisions are automated
- No attempt is made to identify individuals from the data
